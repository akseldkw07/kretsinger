```python
import sys

# Get all imported modules
imported_modules = sys.modules.keys()

# Filter to show only the main packages (not submodules)
main_packages = sorted([mod.split(".")[0] for mod in imported_modules if not mod.startswith("_")])

# Remove duplicates
main_packages = sorted(set(main_packages))

print("Imported packages:")
for pkg in main_packages:
    print(f"  {pkg}")
```

# Sandbox

```python
import sys

# All imported modules
imported_modules = set(sys.modules.keys())

# Standard library modules (Python 3.10+)
stdlib_modules = set(sys.stdlib_module_names)

# Extract top-level package names
top_level = {mod.split(".")[0] for mod in imported_modules if not mod.startswith("_")}

# Third-party packages = imported minus stdlib
third_party_packages = sorted(pkg for pkg in top_level if pkg not in stdlib_modules)

print("Third-party packages:")
for pkg in third_party_packages:
    print(f"  {pkg}")
```
