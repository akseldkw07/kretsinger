# With Inheritance

```python
@post_init
class TestClass:
    arg1: str
    arg2: int
    arg2_sq: int

    def __init__(self, arg1: str, arg2: int) -> None:
        self.arg1 = arg1
        self.arg2 = arg2

    def __post_init__(self) -> None:
        self.arg2_sq = self.arg2**2
```

```python
testclass = TestClass("hello", 5)
```

```python
testclass.arg2_sq
```

```python
class TestChild(TestClass):
    arg3: float
    arg3_sqrt: float

    def __init__(self, arg1: str, arg2: int, arg3: float) -> None:
        super().__init__(arg1, arg2)
        self.arg3 = arg3

    def __post_init__(self) -> None:
        super().__post_init__()
        self.arg3_sqrt = self.arg3**0.5
```

```python
testchild = TestChild("world", 9, 16.0)
```

```python
testchild.arg2_sq
```

```python
testchild.arg3_sqrt
```
