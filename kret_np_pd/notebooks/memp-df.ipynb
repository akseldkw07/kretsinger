## Test Class

```python
class MyInputDict(InputTypedDict):
    aux: pd.DataFrame
```

```python
class MyMemoDataFrame(MemoDataFrame[MyInputDict]):
    @memo_array
    def a_sq(self):
        return t.cast(np.ndarray, (self.data["a"] ** 2).to_numpy())

    @memo_array
    def compute_sum(self):
        return t.cast(np.ndarray, (self.data.sum(axis=1) + self.inputs["aux"].sum(axis=1)).to_numpy())
```

```python
memo2 = MyMemoDataFrame({"data": df, "aux": aux})
```
